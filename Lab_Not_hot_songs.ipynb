**Prepare a new dataset of not_hot_songs. This dataset can contain songs of your choice, others collected from the web or any other combination**

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 2. Explore the dataset

songs_file = pd.read_csv('Spotify_Top_Hit_Playlist_2010_2022.csv')
songs_file.head()


# Dataset taken from: https://www.kaggle.com/datasets/josephinelsy/spotify-top-hit-playlist-2010-2022
```

```python
# Create a new DataFrame with columns "track_name" as Title an "artist_name" as Artist

not_hot_song = pd.DataFrame().assign(Titles=songs_file['track_name'], Artist=songs_file['artist_name'])
not_hot_song.head()
```

```python
#Save the DataFrame

not_hot_song.to_csv('not_hot_song.csv', index=False)
```
